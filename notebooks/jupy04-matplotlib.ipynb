
````python
# Data
time = range(0, 100)
position = np.cumsum(np.random.normal(loc=0.0, scale=1.0, size=100))

# Plot it with a lineplot
plt.plot(time, position)
plt.show()
#plt.savefig(fname="my_first_fig.png", dpi=300, bbox_inches="tight")
````

````python
# Plot it with a lineplot
plt.plot(time, position, color="red", linestyle="dashed", linewidth=0.75)
plt.xlabel(xlabel="Time")
plt.ylabel(ylabel="Position")
plt.title(label="Brownian motion (aka drunkard walk)")
plt.show()
````

````python
# Data
x = range(0, 100)
y1 = np.cumsum(np.random.normal(loc=0.0, scale=1.0, size=100))
y2 = np.cumsum(np.random.normal(loc=0.0, scale=1.0, size=100))

# Create the figure and two subplots (1 row, 2 columns)
fig, ax = plt.subplots(nrows=1, ncols=2)

# Plot on the first subplot (ax[0])
ax[0].plot(x, y1, color="blue", linestyle="dashed")
ax[0].set_title("The second random walk")

# Plot on the second subplot (ax[1])
ax[1].plot(x, y2, color="red", linestyle="dashed")
ax[1].set_title("The second random walk")

# Display the plot
fig.tight_layout()  # Adjust spacing between subplots
#fig.show()
````

````python
penguin_count = palmer.groupby("species").size()
print(penguin_count)

plt.bar(x=penguin_count.index, height=penguin_count.values, color="rebeccapurple")
plt.xlabel("Penguin species")
plt.ylabel("Number of individuals")
plt.title("Penguin counts in the Palmer penguins dataset by species")
plt.tight_layout()
plt.show()
````

````python
# Group by species and compute mean and std
grouped = palmer.groupby("species")["body_mass_g"]
mean_vals = grouped.mean()
sem_vals = grouped.sem()  # or use .sem() for standard error

# Plot with error bars
plt.bar(x=mean_vals.index, height=mean_vals.values, yerr=sem_vals.values, capsize=5, color="lightcoral")
plt.ylabel("Bill Length (mm)")
plt.title("Average body mass (in g) of penguins by species (with sem)")
plt.grid(visible=True, axis="y", linestyle="dashed", alpha=0.7)
plt.show()
````

````python
import matplotlib.pyplot as plt

# Drop missing values (optional but recommended for clean plotting)
#data = palmer.dropna(subset=["species", "sex", "body_mass_g"])

# Group by species and sex
penguins_grouped = palmer.dropna().groupby(["species", "sex"])["body_mass_g"]
mean_vals = penguins_grouped.mean()
sem_vals = penguins_grouped.sem()

# Flatten multi-index for plotting
mean_vals = mean_vals.unstack()
sem_vals = sem_vals.unstack()

# Plot grouped bar chart
species = mean_vals.index
x = range(len(species))
bar_width = 0.35

plt.bar(
    x=[i - bar_width/2 for i in x],
    height=mean_vals["male"],
    yerr=sem_vals["male"],
    width=bar_width,
    label="male",
    capsize=5,
    color="sandybrown"
)

plt.bar(
    x=[i + bar_width/2 for i in x],
    height=mean_vals["female"],
    yerr=sem_vals["female"],
    width=bar_width,
    label="female",
    capsize=5,
    color="mediumseagreen"
)

# Labels and legend
plt.xticks(x, species)
plt.ylabel("Body Mass (g)")
plt.title("Average Body Mass by Species and Sex (with sem added)")
plt.grid(visible=True, axis="y", linestyle="dashed", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Load and clean the data
palmer = pd.read_csv("data/penguins.csv")
palmer = palmer.dropna(subset=["bill_length_mm", "bill_depth_mm", "species"])

# Convert species to category codes
species_cat = pd.Categorical(palmer["species"])

# Define the colors manually (corresponding to the category codes)
color_map = {"Adelie": "red", "Chinstrap": "blue", "Gentoo": "green"}

# Plot using raw codes as colors (with manual mapping to colors)
plt.figure(figsize=(8, 6))
plt.scatter(
    x=palmer["bill_length_mm"],
    y=palmer["bill_depth_mm"],
    c=[color_map[species_cat.categories[code]] for code in species_cat.codes],  # Apply the color map
    alpha=0.7
)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.title("Penguin Bill Dimensions by Species")

# Create a legend using the manually defined color map
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=species_cat.categories[i],
           markerfacecolor=color_map[species_cat.categories[i]], markersize=8)
    for i in range(len(species_cat.categories))
]
plt.legend(handles=legend_elements, title="Species")

plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
````
````python
import numpy as np

# Load the data
palmer = pd.read_csv(filepath_or_buffer="data/penguins.csv")

# Calculate the mean of 'body_mass_g' grouped by 'species' and 'sex'
mean_bill_length = palmer.groupby(['species', 'sex'])['body_mass_g'].mean()

# Calculate the standard error of the mean (SEM) for the bill length
sem_bill_length = palmer.groupby(['species', 'sex'])['body_mass_g'].sem()

# Unstack to get the data in a format suitable for plotting
mean_bill_length = mean_bill_length.unstack(fill_value=np.nan)
sem_bill_length = sem_bill_length.unstack(fill_value=np.nan)

# Plotting the bar chart with error bars
ax = mean_bill_length.plot(kind="bar", yerr=sem_bill_length, capsize=5, figsize=(10, 6), color=["skyblue", "hotpink"])

# Customizing the plot
plt.title("Mean body mass in penguins on Palmer Archipelago by species and sex")
plt.xlabel("Species")
plt.ylabel("Mean body mass (in g)")
plt.xticks(rotation=0.0)  # Ensure the species names are readable
plt.legend(title="Sex", labels=["Male", "Female"])

# Show the plot
plt.tight_layout()
plt.show()
````

````python
def send_email(to, **kwargs):
    subject = kwargs.get('subject', 'No Subject')
    body = kwargs.get('body', 'Hello, this is your message.')
    attachments = kwargs.get('attachments', [])
    
    print(f"Sending email to {to}")
    print(f"Subject: {subject}")
    print(f"Body: {body}")
    if attachments:
        print(f"Attachments: {attachments}")
    else:
        print("No attachments.")
        
# This works dynamically without needing to modify the function if new keys are added
send_email('john@example.com', subject='Meeting', body='See you at 3', attachments=['file1.pdf'])
````

````python
import this
#print(this.s)
print(this.d)
````

````python
def rot13(text):
    return ''.join(this.d.get(char, char) for char in text)


print(rot13("I like Python"))

# Example usage
original_text = "Hello World!"
encoded_text = rot13(original_text)
decoded_text = rot13(encoded_text)

print(encoded_text)
print(decoded_text)
````
